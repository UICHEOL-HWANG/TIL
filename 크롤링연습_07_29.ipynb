```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

url = "https://movie.naver.com/movie/point/af/list.naver"
html = requests.get(url)
soup = bs(html.text, "html.parser")

titles = soup.select("tbody")

titles[0].find_all('tr')[0].text.split('\n')[1]
```

```python
titles[0].find_all('tr')[0].text.split('\n')[3]
```

```python
titles[0].find_all("tr")[0].text.split("\n")[5].split("점 중")[1]
```

```python
titles[0].find_all('tr')[0].text.split('\n')[7]
```

```python
title_len = len(titles[0].find_all('tr')) 
df = pd.DataFrame(columns= ['num','name','score','review'])
for i in range(0,title_len):
  num = titles[0].find_all('tr')[i].text.split('\n')[1]
  name = titles[0].find_all('tr')[i].text.split('\n')[3]
  score = titles[0].find_all("tr")[i].text.split("\n")[5].split("점 중")[1]
  review = titles[0].find_all('tr')[i].text.split('\n')[7]
  df = df.append({'num':num,'name':name,'score':score,'review':review},ignore_index=True)
```

```python
def collection__review(x):
    df = pd.DataFrame(columns=["번호", "제목", "평점", "감상평"])
    url = f"https://movie.naver.com/movie/point/af/list.naver?&page={x}"
    table_pd = pd.read_html(url)
    for i in range(10):
      try:
        index = table_pd[0]['번호'][i]
        name = table_pd[0]['감상평'][i].split("별점 - 총 10점 중")[0]
        data = table_pd[0]["감상평"][i].split("  별점 - 총 10점 중")[1][:-3].split("  ")
      except:
        pass
        if len(data) == 2:
          df = df.append({"번호":int(index), "제목":name, "평점":int(data[0]), "감상평":data[1]}, ignore_index=True)
    return df
```

```python
naver_url = "https://kin.naver.com/search/list.naver?query=%EC%9E%AC%EB%8A%A5%EA%B8%B0%EB%B6%80"
html_1 = requests.get(naver_url)
soup = bs(html_1.text, "html.parser")

title = soup.select("#s_content > div.section > ul")

n_title = title[0].find_all('li')
```

```python
df = pd.DataFrame(columns=["제목", "등록일","내용"])


for i in  n_title:
  df.append({"제목":i.find('dt').text,"등록일":i.find_all("dd")[0].text,"내용":i.find_all("dd")[1].text},ignore_index=True)
  
df
```
